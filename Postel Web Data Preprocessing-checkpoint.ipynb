## 1. Data Understanding

```python
import pandas as pd
```

```python
# Load CSV file
file_path = 'scraped_data3.csv'  
df = pd.read_csv(file_path)

print(df.dtypes)
```

```python
df.describe()
```

## 2. Data Cleaning

### 2.1. Konversi Tanggal ke Datetime

```python
df['Issued Date'] = pd.to_datetime(df['Issued Date'], format='%d %b %Y', errors='coerce')
```

```python
print(df.dtypes)
```

```python
df.iloc[1000:1100]
```

### 2.2. Hapus Duplikat Data (Certificate Number)

```python
df['Certificate Number'] = df['Certificate Number'].str.strip()
df['Certificate Number'] = df['Certificate Number'].str.lower()

# Identify duplicates based on 'Certificate Number' after cleaning
duplicates = df[df.duplicated(subset='Certificate Number', keep=False)]

print(f"Banyaknya data duplikat {len(duplicates)}")
```

```python
from IPython.display import display

sorted_duplicates = duplicates.sort_values(by='Certificate Number')

display(sorted_duplicates.head(10))
```

```python
df = df.drop_duplicates(subset='Certificate Number')
```

```python
row_count = len(df)
print(f"Jumlah data: {row_count}")
```

### 2.3. Cari Nilai NaN setiap kolom

```python
nan_counts = df.isna().sum()

# Display the number of NaN values in each column
print("Jumlah nilai NaN setiap kolom")
print(nan_counts)
```

```python
nan_applicant_rows = df[df['Name Of Applicant'].isna()]

display(nan_applicant_rows)
```

```python
nan_applicant_rows = df[df['Marketing Name'].isna()]

display(nan_applicant_rows)
```

```python
nan_applicant_rows = df[df['Certificate Number'].isna()]

display(nan_applicant_rows)
```

```python
df['Certificate Number'].fillna('0000/postel/2017', inplace=True)
df['Name Of Applicant'].fillna('KOMINFO', inplace=True)
```

### 2.4. Normalisasi Nama Merk

```python
merk_counts = df['Merk'].value_counts().sort_values(ascending=False)
print(merk_counts.head(20))
```

```python
replacements = {
    'hp': 'HP',
    'ACER': 'Acer',
    'LENOVO': 'Lenovo',
    'CISCO' : 'Cisco',
    'HUAWEI' : 'Huwaei',
    'Cisco Systems' : 'Cisco',
    'SAMSUNG' : 'Samsung'
}

df['Merk'] = df['Merk'].replace(replacements)
```

```python
merk_counts = df['Merk'].value_counts().sort_values(ascending=False)
print(merk_counts.head(20))
```

## 3. Visualisasi Data

```python
import matplotlib.pyplot as plt

df['Year'] = df['Issued Date'].dt.year
yearly_permit_counts = df['Year'].value_counts().sort_index()

# Create a bar chart
plt.figure(figsize=(10, 6))
yearly_permit_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Permits Issued by Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Permits', fontsize=14)
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.grid(axis='y')
```
